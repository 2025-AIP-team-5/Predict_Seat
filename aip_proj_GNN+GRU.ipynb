### Dataset preprocessing

```python
# column 'index' => 'time'
df = df.rename(columns={"index": "time"})
df = df.set_index("time")

# remove 'NaN'
df = df.apply(pd.to_numeric, errors='coerce').fillna(0).astype(int)

```

```python
seat_matrix = df.T
columns_to_drop = [col for col in seat_matrix.columns if str(col).isdigit() and int(col) >= 1300]
seat_matrix = seat_matrix.drop(columns=columns_to_drop)

seat_matrix
```

### Calculate similarity betweens seats

```python
from sklearn.metrics import jaccard_score
from itertools import combinations

similarities = []
seat_ids = seat_matrix.index.tolist()

for a, b in combinations(seat_ids, 2):  # 좌석 쌍 조합
    try:
        sim = jaccard_score(seat_matrix.loc[a], seat_matrix.loc[b])
        if sim > 0:  # 0 이상인 경우만 저장
            similarities.append((a, b, sim))
    except ValueError:
        # 모든 값이 0인 경우 (계산 불가능), 무시
        continue

# 유사도 결과를 DataFrame으로 저장
sim_df = pd.DataFrame(similarities, columns=["seat_a", "seat_b", "jaccard_similarity"])

# 유사도 높은 순으로 정렬
sim_df_sorted = sim_df.sort_values(by="jaccard_similarity", ascending=False)

# 상위 30개 출력 (원하면 전체 저장도 가능)
print(sim_df_sorted.head(10))
```

### Generate Graph

```python
def spatial_similarity(a, b):
    try:
        a_ten = int(str(a).zfill(2)[0])
        b_ten = int(str(b).zfill(2)[0])
        diff = abs(a_ten - b_ten)
        if diff == 0:
            return 1.0
        elif diff == 1:
            return 0.5
        else:
            return 0.0
    except:
        return 0.0
```

```python
seat_list = sorted(set(sim_df_sorted["seat_a"]).union(sim_df_sorted["seat_b"]))
seat_to_idx = {seat: i for i, seat in enumerate(seat_list)}

edge_index = []
edge_weights = []

for a, b, jaccard in sim_df_sorted[["seat_a", "seat_b", "jaccard_similarity"]].values:
    i = seat_to_idx[a]
    j = seat_to_idx[b]
    spatial = spatial_similarity(a, b)
    weight = alpha * spatial + beta * jaccard

    edge_index.append([i, j])
    edge_index.append([j, i])  # 반대 방향도 추가
    edge_weights.append(weight)
    edge_weights.append(weight)  # 같은 가중치 사용 (대칭)
```

```python
import numpy as np

# seat_matrix: (seat_id, full time series) — 예: (48, 6×4=24)
num_series = 4
time_steps_per_series = 6  # 예: 1000, 1030, ..., 1230

seat_matrix_aligned = seat_matrix.loc[seat_list]  # 좌석 순서 정렬

X_list, Y_list = [], []
for i in range(num_series):
    start = i * 6
    end = start + 6
    series = seat_matrix_aligned.iloc[:, start:end]

    for t in range(5):  # 6개 시간대니까 5쌍 생성
        X_list.append(series.iloc[:, t].values)     # (N,)
        Y_list.append(series.iloc[:, t+1].values)   # (N,)

X_np = np.stack(X_list)[:, :, None]  # shape (T, N, 1)
Y_np = np.stack(Y_list)[:, :, None]  # shape (T, N, 1)

X_np
# X_all = []
# Y_all = []

# for i in range(num_series):
#     start = i * time_steps
#     end = start + time_steps
#     series = seat_matrix_aligned.iloc[:, start:end]

#     for t in range(time_steps - 1):
#         X_list.append(series.iloc[:, t].values)
#         Y_list.append(series.iloc[:, t + 1].values)

# # 최종 numpy 형태
# X_np = np.stack(X_all)[:, :, None]  # shape: (T_total, N, 1)
# Y_np = np.stack(Y_all)[:, :, None]  # shape: (T_total, N, 1)
```

```python
import torch
from torch_geometric.nn import GCNConv
from torch.nn import GRU, Linear
import torch.nn.functional as F
from torch_geometric.data import Data

# GCN + GRU 모델 정의
class GCN_GRU(torch.nn.Module):
    def __init__(self, in_channels, hidden_channels, out_channels):
        super().__init__()
        self.gcn = GCNConv(in_channels, hidden_channels)
        self.gru = GRU(hidden_channels, hidden_channels, batch_first=True)
        self.fc = Linear(hidden_channels, out_channels)

    def forward(self, x_seq, edge_index, edge_weight):
        '''
        x_seq: (T, N, F)
        edge_index: (2, E)
        edge_weight: (E,)
        '''
        gcn_outputs = []
        for t in range(x_seq.size(0)):
            x_t = x_seq[t]  # shape: (N, F)
            z_t = F.relu(self.gcn(x_t, edge_index, edge_weight))  # (N, hidden)
            gcn_outputs.append(z_t)

        gcn_outputs = torch.stack(gcn_outputs, dim=0)        # (T, N, hidden)
        gcn_outputs = gcn_outputs.transpose(0, 1)             # (N, T, hidden)

        gru_out, _ = self.gru(gcn_outputs)                    # (N, T, hidden)
        last_out = gru_out[:, -1, :]                          # (N, hidden)
        pred = torch.sigmoid(self.fc(last_out))               # (N, out_channels)
        return pred
```

```python
from sklearn.model_selection import train_test_split

param_grid = {
    "hidden_channels": [16, 32, 64],
    "lr": [0.1, 0.01, 0.001],
    "alpha": [ 1, 0.75, 0.5, 0.25, 0 ],
    "beta": [ 0, 0.25, 0.5, 0.75, 1 ],
}

X_train, X_val, Y_train, Y_val = train_test_split(X_np, Y_np, test_size=0.25, random_state=42)

# 데이터 준비
X_train = torch.tensor(X_train, dtype=torch.float)
Y_train = torch.tensor(Y_train, dtype=torch.float)
X_val = torch.tensor(X_val, dtype=torch.float)
Y_val = torch.tensor(Y_val, dtype=torch.float)

best_val_loss = float("inf")
best_params = None

for hidden in param_grid["hidden_channels"]:
    for lr in param_grid["lr"]:
      for alpha, beta in zip(param_grid["alpha"], param_grid["beta"]):

        # generate graph
        edge_index = []
        edge_weights = []

        for a, b, jaccard in sim_df_sorted[["seat_a", "seat_b", "jaccard_similarity"]].values:
            i = seat_to_idx[a]
            j = seat_to_idx[b]
            spatial = spatial_similarity(a, b)
            weight = alpha * spatial + beta * jaccard

            edge_index.append([i, j])
            edge_index.append([j, i])  # 반대 방향도 추가
            edge_weights.append(weight)
            edge_weights.append(weight)  # 같은 가중치 사용 (대칭)

        edge_index_tensor = torch.tensor(edge_index, dtype=torch.long).T
        edge_weight_tensor = torch.tensor(edge_weights, dtype=torch.float)

        model = GCN_GRU(1, hidden, 1)
        optimizer = torch.optim.Adam(model.parameters(), lr=lr)
        loss_fn = torch.nn.BCELoss()

        # 학습
        for epoch in range(50):
            model.train()
            total_loss = 0
            for i in range(len(X_train)):
                x_seq = X_train[i].unsqueeze(0).repeat(5, 1, 1)
                y = Y_train[i]
                pred = model(x_seq, edge_index_tensor, edge_weight_tensor)
                loss = loss_fn(pred, y)
                optimizer.zero_grad()
                loss.backward()
                optimizer.step()
                total_loss += loss.item()

        # 검증
        model.eval()
        val_loss = 0
        with torch.no_grad():
            for i in range(len(X_val)):
                x_seq = X_val[i].unsqueeze(0).repeat(5, 1, 1)
                y = Y_val[i]
                pred = model(x_seq, edge_index_tensor, edge_weight_tensor)
                loss = loss_fn(pred, y)
                val_loss += loss.item()

        avg_val_loss = val_loss / len(X_val)
        print(f"[hidden={hidden}, lr={lr}, alpha={alpha}, beta={beta}] → val_loss = {avg_val_loss:.4f}")

        if avg_val_loss < best_val_loss:
            best_val_loss = avg_val_loss
            best_params = {"hidden_channels": hidden, "lr": lr, "alpha": alpha, "beta": beta}

print("Best Hyperparameters:", best_params)
```
